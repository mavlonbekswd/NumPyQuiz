Quiz 07 — arange, linspace, zeros, ones  [SOLVED]
Theme  : Taxi Rides

```python
import numpy as np

# arange(start, stop, step) — stop is EXCLUSIVE
distances = np.arange(1, 21, 1)
print("Distances  :", distances)

# linspace(start, stop, num) — stop is INCLUSIVE, num = count of values
fare_tiers = np.linspace(3.0, 30.0, 10)
print("Fare tiers :", fare_tiers)

# zeros / ones take a shape tuple or integer
ratings = np.zeros(5)
print("Ratings    :", ratings)

drivers = np.ones(5)
print("Drivers    :", drivers)

# ── WHY ────────────────────────────────────────────────────────────────────
# arange is best when you know the step size.
# linspace is best when you know the number of points.
# zeros / ones initialise placeholder arrays before filling with real data.
```
