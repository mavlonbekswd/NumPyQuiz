Quiz 13 — Stacking + Joining  [SOLVED]
Theme  : Travel Bookings

```python
import numpy as np

agency_a = np.array([45, 38, 52, 60, 70, 85, 90])
agency_b = np.array([30, 42, 55, 48, 65, 78, 88])

# concatenate joins along an existing axis (default axis=0 for 1-D = end-to-end)
combined = np.concatenate([agency_a, agency_b])
print("Combined (14,) :", combined)

# vstack treats each 1-D array as a row
stacked = np.vstack([agency_a, agency_b])
print("Vstack shape   :", stacked.shape)    # (2, 7)

# hstack on 1-D arrays behaves identically to concatenate
print("Hstack (14,)   :", np.hstack([agency_a, agency_b]))

# sum down the rows (axis=0) gives total per day column
print("Daily totals   :", stacked.sum(axis=0))

# ── WHY ────────────────────────────────────────────────────────────────────
# vstack is the clearest way to build a matrix from rows.
# For 1-D arrays hstack == concatenate; for 2-D they differ.
# axis=0 sum collapses rows, leaving one value per column (per day).
```
