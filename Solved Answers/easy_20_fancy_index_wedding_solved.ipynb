Quiz 20 — Fancy Indexing  [SOLVED]
Theme  : Wedding Seating

```python
import numpy as np

guest_names   = ["Alice","Bob","Carol","Dan","Eve","Frank","Grace","Hank","Ivy","Jack"]
table_numbers = np.array([3, 1, 2, 5, 1, 4, 2, 3, 5, 1])
meal_choice   = np.array(["Veg","Fish","Meat","Veg","Meat","Fish","Veg","Meat","Fish","Veg"])

# Fancy indexing: pass a list/array of indices → returns values at those positions
idx = [1, 4, 8]
print("Tables for guests 2,5,9   :", table_numbers[idx])
print("Meals for guests 2,5,9    :", meal_choice[idx])

# np.where with one argument returns a tuple of index arrays (one per dimension)
table1_idx = np.where(table_numbers == 1)[0]
print("Table 1 guest indices     :", table1_idx)

# Use those indices to look up names from a Python list
table1_names = [guest_names[i] for i in table1_idx]
print("Table 1 guest names       :", table1_names)

# ── WHY ────────────────────────────────────────────────────────────────────
# Fancy indexing (integer array index) creates a COPY, unlike slice views.
# np.where(condition) returns a tuple — [0] extracts the 1-D index array.
# This pattern (np.where → indices → lookup) is extremely common in data work.
```
