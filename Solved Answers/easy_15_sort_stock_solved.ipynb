Theme  : Stock / Sales Simulation

```python
import numpy as np

prices = np.array([145.2, 138.7, 152.4, 149.1, 135.0,
                   160.3, 142.8, 155.9, 148.6, 133.4])

# np.sort returns a new sorted array (does not modify original)
asc = np.sort(prices)
print("Ascending  :", asc)

# Reverse the sorted array for descending order
desc = np.sort(prices)[::-1]
print("Descending :", desc)

# argmax / argmin return the INDEX of the extreme value
hi_idx = np.argmax(prices)
lo_idx = np.argmin(prices)
print(f"Highest day index : {hi_idx}  (price: {prices[hi_idx]})")
print(f"Lowest day index  : {lo_idx}  (price: {prices[lo_idx]})")

# ── WHY ────────────────────────────────────────────────────────────────────
# np.sort does not sort in place — use arr.sort() if you want in-place.
# [::-1] reverses any array/sequence efficiently.
# argmax/argmin are more useful than max/min when you need the position,
# not just the value — e.g. to look up a name, date, or other attribute.
```
