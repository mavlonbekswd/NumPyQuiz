Quiz 34 — np.where + Multi-condition Labelling  [SOLVED]
Theme  : Customer Support Tickets

```python
import numpy as np

times    = np.array([2, 48, 5, 72, 1, 24, 96, 3, 36, 8, 120, 4, 18, 60, 12])
priority = np.array([1,  2, 1,  3, 1,  2,  3, 1,  2, 1,   3, 2,  2,  3,  1])

# SLA limits per priority
sla_limit = np.where(priority == 1, 4, np.where(priority == 2, 24, 72))

# np.select: list of conditions → list of choices (first True condition wins)
conditions = [times <= sla_limit]
choices    = ["Met SLA"]
status     = np.select(conditions, choices, default="Breached SLA")
print("SLA Status:", status)

# Breached mask
breached = status == "Breached SLA"
print(f"\nTotal breached: {breached.sum()}")
for p in [1, 2, 3]:
    count = ((priority == p) & breached).sum()
    print(f"  Priority {p} breached: {count}")

# Hours over SLA for breached tickets
over_hours = np.where(breached, times - sla_limit, 0)

# Print breached ticket details
print("\nBreached ticket details:")
for i in np.where(breached)[0]:
    print(f"  Ticket {i+1}: P{priority[i]}  Time={times[i]}h  "
          f"SLA={sla_limit[i]}h  Over by {over_hours[i]}h")

# ── WHY ────────────────────────────────────────────────────────────────────
# Nested np.where builds a vectorised if/elif/else across the array.
# np.select is cleaner for 3+ conditions — first matching condition wins.
# over_hours uses np.where to zero-out non-breached tickets cleanly.
```
