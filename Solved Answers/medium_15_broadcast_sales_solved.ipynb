Quiz 35 — Advanced Broadcasting  [SOLVED]
Theme  : Sales Simulation (Multi-region)

```python
import numpy as np

np.random.seed(14)
sales       = np.random.randint(50, 500, (3, 4, 5))
prices      = np.array([10, 25, 50, 80])
region_mult = np.array([1.0, 1.15, 0.90])

# sales: (3,4,5)   prices: (4,) → reshape to (1,4,1) to broadcast over (3,4,5)
# (1,4,1) × (3,4,5) → (3,4,5)  — prices repeat across regions and months
revenue = sales * prices[np.newaxis, :, np.newaxis]
print("Revenue shape:", revenue.shape)

# region_mult: (3,) → reshape to (3,1,1)
# (3,1,1) × (3,4,5) → (3,4,5)
revenue = revenue * region_mult[:, np.newaxis, np.newaxis]

# Sum over products (axis=1) and months (axis=2)
total_per_region = revenue.sum(axis=(1, 2))
print("Total per region:", np.round(total_per_region, 0))

# Sum over months first → (3,4), then argmax
rev_by_region_product = revenue.sum(axis=2)      # (3,4)
flat_idx = np.argmax(rev_by_region_product)
r, p     = np.unravel_index(flat_idx, (3, 4))
print(f"Best region-product: region={r}, product={p}")

# ── WHY ────────────────────────────────────────────════════════════════════
# np.newaxis inserts a size-1 axis, enabling broadcasting alignment.
# prices shape: (4,) → (1,4,1) aligns with (3,4,5):
#   dim 0: 1 broadcasts to 3 (regions)
#   dim 1: 4 matches 4 (products)
#   dim 2: 1 broadcasts to 5 (months)
# This avoids any explicit loop across regions or months.
```
