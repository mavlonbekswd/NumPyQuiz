Quiz 36 — Variance + Volatility  [SOLVED]
Theme  : Stock Prices

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(17)
stock_a = np.random.normal(150, 5, 30)
stock_b = np.random.normal(150, 20, 30)

for name, s in [("Stock A", stock_a), ("Stock B", stock_b)]:
    print(f"{name} — Mean: {s.mean():.2f}  Var: {s.var():.2f}  Std: {s.std():.2f}")

# Daily % change using array slicing — no loop
# price[1:] are "today's" prices; price[:-1] are "yesterday's"
pct_a = (stock_a[1:] - stock_a[:-1]) / stock_a[:-1] * 100
pct_b = (stock_b[1:] - stock_b[:-1]) / stock_b[:-1] * 100

print(f"\nVolatility (std of daily % change):")
print(f"  Stock A: {pct_a.std():.2f}%")
print(f"  Stock B: {pct_b.std():.2f}%")
winner = "Stock B" if pct_b.std() > pct_a.std() else "Stock A"
print(f"  Higher volatility: {winner}")

days = np.arange(1, 31)
plt.figure(figsize=(10, 5))
plt.plot(days, stock_a, label="Stock A (low vol)", color="steelblue")
plt.plot(days, stock_b, label="Stock B (high vol)", color="tomato")
plt.title("Stock Price Comparison — 30 Days")
plt.xlabel("Day")
plt.ylabel("Price (£)")
plt.legend()
plt.tight_layout()
plt.savefig("medium_16_stock_prices.png", dpi=100)
plt.show()

# ── WHY ────────────────────────────────────────────────────────────────────
# price[1:] - price[:-1] computes differences between consecutive elements
# without any loop — this is the vectorised "diff" pattern.
# Variance = std² — std is more interpretable (same units as the data).
```
