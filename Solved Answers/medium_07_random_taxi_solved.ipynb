Quiz 27 — Random: choice, shuffle, permutation  [SOLVED]
Theme  : Taxi Ride Simulation

```python
import numpy as np

np.random.seed(99)
drivers = np.array(["D01","D02","D03","D04","D05","D06","D07","D08"])
zones   = np.array(["North","South","East","West","Central"])
fares   = np.array([8.50, 12.00, 6.50, 15.00, 9.00, 11.50, 7.00, 13.00])

# choice(array, size, replace) — picks random elements
zone_assignments = np.random.choice(zones, size=8, replace=True)
print("Zone assignments :", zone_assignments)

# permutation returns a NEW shuffled array — original is unchanged
ride_order = np.random.permutation(len(drivers))
print("Ride order (perm):", ride_order)

# shuffle modifies IN-PLACE — no return value
fares_copy = fares.copy()           # preserve original
np.random.shuffle(fares_copy)
print("Shuffled fares   :", fares_copy)

# Simulate 20 fares from original list
sim_fares = np.random.choice(fares, size=20, replace=True)
print(f"20 draws mean: {sim_fares.mean():.2f}  std: {sim_fares.std():.2f}")

# ── WHY ────────────────────────────────────────────────────────────────────
# shuffle modifies in-place and returns None — a common mistake is to write
# fares = np.random.shuffle(fares) which sets fares to None.
# permutation is safer when you need the shuffled result without destroying
# the original order.
```
