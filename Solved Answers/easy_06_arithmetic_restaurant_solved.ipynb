Quiz 06 — Arithmetic Operations  [SOLVED]
Theme  : Restaurant Bills
Topics : element-wise operations, np.round, single-element update

```python
import numpy as np

food   = np.array([45.5, 30.0, 62.0, 18.5, 55.0, 40.0, 72.5, 25.0])
drinks = np.array([12.0,  8.5, 20.0,  5.0, 15.5, 10.0, 22.0,  7.5])

# Element-wise addition — no loop needed
total = food + drinks
print("Total bills    :", total)

# Multiply every element by 1.10, then round
with_service = np.round(total * 1.10, 2)
print("After service  :", with_service)

# Update single element: table 5 = index 4
with_service[4] -= 5
print("After discount :", with_service)

# ── WHY ────────────────────────────────────────────────────────────────────
# NumPy applies arithmetic operators element-by-element when shapes match.
# This is far faster than a Python for-loop for large arrays.
# np.round(arr, 2) rounds every element to 2 decimal places in one call.
```
