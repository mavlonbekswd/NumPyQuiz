Quiz 23 — Broadcasting + Reshape  [SOLVED]
Theme  : Hotel Room Pricing

```python
import numpy as np

rates       = np.array([80, 120, 200, 350])
multipliers = np.array([1.0, 1.1, 1.2, 1.3, 1.5])

# reshape(4,1): shape (4,1) × shape (5,) → NumPy pads (5,) to (1,5)
# result: (4,1) × (1,5) = (4,5) — each row is one room, each col one night
price_matrix = np.round(rates.reshape(4, 1) * multipliers, 2)
print("Price matrix (4 rooms × 5 nights):")
print(price_matrix)

# unravel_index converts flat index to (row, col)
flat_max   = np.argmax(price_matrix)
room_i, night_i = np.unravel_index(flat_max, price_matrix.shape)
print(f"Most expensive: Room {room_i}, Night {night_i}, £{price_matrix[room_i, night_i]}")

print("Total revenue : £", price_matrix.sum())

# ── WHY ────────────────────────────────────────────────────────────────────
# Broadcasting rule (right-align dimensions):
#   rates    : (4, 1)
#   multipliers: (   5)  → treated as (1, 5)
#   result   : (4, 5)
# The (4,1) column repeats across 5 columns; (1,5) row repeats across 4 rows.
# np.unravel_index is the standard way to get row+col from a flat argmax.
```
