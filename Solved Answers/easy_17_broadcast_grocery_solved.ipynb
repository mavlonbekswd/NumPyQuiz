Quiz 17 — Broadcasting  [SOLVED]
Theme  : Grocery Discounts

```python
import numpy as np

prices = np.array([[1.20, 2.50, 0.80, 3.00],
                   [4.50, 5.20, 3.80, 6.00],
                   [0.60, 1.10, 0.90, 1.50]])

# Scalar 0.90 broadcast across all 12 elements simultaneously
discounted_all = np.round(prices * 0.90, 2)
print("10% off (all)   :")
print(discounted_all)

# Category discounts: shape (3,) → must be (3,1) to broadcast over (3,4)
# (3,) would broadcast across columns; (3,1) broadcasts across rows
rates = np.array([0.95, 0.85, 0.92]).reshape(3, 1)  # shape (3,1)
discounted_cat = np.round(prices * rates, 3)
print("Category discounts:")
print(discounted_cat)

# ── WHY ────────────────────────────────────────────────────────────────────
# Broadcasting rule: dimensions are aligned from the RIGHT.
# prices (3,4) × rates (3,1) → rates column repeats 4 times → (3,4) result.
# rates (3,) would align as (1,3) not (3,1) and produce a shape mismatch.
# reshape(3,1) is the key step that makes per-row broadcasting work.
```
