Quiz 29 — Standard Deviation + Variance  [SOLVED]
Theme  : Restaurant Bills

```python
import numpy as np

np.random.seed(5)
bills = np.clip(np.random.normal(55, 18, 20), 10, 120)

mean   = bills.mean()
median = np.median(bills)
std    = bills.std()
var    = bills.var()

print(f"Mean: {mean:.2f}  Median: {median:.2f}  Std: {std:.2f}  Var: {var:.2f}")

# CV measures relative spread — useful for comparing datasets with different scales
cv = (std / mean) * 100
print(f"CV: {cv:.1f}%")

# Within 1 SD
within_1sd = bills[(bills >= mean - std) & (bills <= mean + std)]
pct = len(within_1sd) / len(bills) * 100
print(f"Within 1 SD: {len(within_1sd)} bills ({pct:.0f}%)")

# Outlier candidates: more than 2 SDs from mean
outliers = bills[np.abs(bills - mean) > 2 * std]
print("Outlier candidates:", np.round(outliers, 2))

# ── WHY ────────────────────────────────────────────────────────────────────
# std (standard deviation) measures average distance from the mean.
# For a perfect normal distribution ~68% of values fall within ±1 SD.
# 2 SD rule catches ~95% of values — anything beyond is a statistical outlier.
# np.abs(bills - mean) computes distances without caring about direction.
```
