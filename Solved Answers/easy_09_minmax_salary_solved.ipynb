Quiz 09 — min, max, sum, cumsum  [SOLVED]
Theme  : Employee Salaries

```python
import numpy as np

salaries = np.array([2800, 3200, 4500, 2200, 5100, 3800, 2600, 4200, 3600, 4900])

print("Min salary   :", np.min(salaries))      # 2200
print("Max salary   :", np.max(salaries))      # 5100
print("Total payroll:", np.sum(salaries))      # 36900

# ptp = "peak to peak" = max - min
print("Salary range :", np.ptp(salaries))      # 2900

# cumsum[i] = sum of all elements from index 0 to i inclusive
print("Cumulative   :", np.cumsum(salaries))

# ── WHY ────────────────────────────────────────────────────────────────────
# np.cumsum is useful for running totals — e.g. tracking budget spend over time.
# np.ptp gives the spread of the data in a single call.
```
