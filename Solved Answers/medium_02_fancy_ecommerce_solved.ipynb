Quiz 22 — Fancy Indexing + Sorting + argsort  [SOLVED]
Theme  : E-commerce Orders

```python
import numpy as np

orders = np.array([[1001, 3, 45.0, 4],
                   [1002, 1, 12.5, 5],
                   [1003, 5, 89.0, 3],
                   [1004, 2, 30.0, 4],
                   [1005, 8,120.0, 2],
                   [1006, 1, 15.0, 5],
                   [1007, 4, 67.5, 4],
                   [1008, 6, 95.0, 3]], dtype=float)

# argsort returns INDICES that would sort the array
price_order = np.argsort(orders[:, 2])          # sort by column 2 (price)
sorted_by_price = orders[price_order]           # fancy-index rows in sorted order
print("Orders sorted by price (asc):")
print(sorted_by_price)

# Last 3 rows in ascending sort = top 3 most expensive
print("\nTop 3 most expensive:")
print(sorted_by_price[-3:])

# Descending: argsort then reverse, OR negate and argsort
rating_order_desc = np.argsort(-orders[:, 3])   # negate to flip order
print("\nOrders by best rating:")
print(orders[rating_order_desc])

# Boolean mask on column 3 (Rating)
high_mask = orders[:, 3] >= 4
print("\nHigh-rated order IDs:", orders[high_mask, 0].astype(int))

# ── WHY ────────────────────────────────────────────────────────────────────
# argsort returns positions, not values — this lets us reorder ALL columns
# of a matrix based on one column (a very common data-wrangling pattern).
# Negating before argsort is efficient for descending sort without copying.
# orders[mask, 0] selects column 0 only from the filtered rows.
```
