Quiz 31 — Mean vs Median + Outlier Detection  [SOLVED]
Theme  : YouTube Analytics

```python
import numpy as np
import matplotlib.pyplot as plt

views = np.array([45,52,48,61,55,58,49,53,60,57,54,50,
                  46,59,51,56,1200,62,47,63,44,55,58,52])

mean_all   = views.mean()
median_all = np.median(views)
print(f"With outlier  — Mean: {mean_all:.1f}k  Median: {median_all:.1f}k")
print("→ Mean is pulled up by the viral month; median is robust.")

# Z-score: how many standard deviations each point is from the mean
z_scores  = (views - mean_all) / views.std()
outlier_idx = np.where(np.abs(z_scores) > 2)[0]
print("Outlier month indices:", outlier_idx)

# Remove outliers and recompute
clean = np.delete(views, outlier_idx)
print(f"Clean mean: {clean.mean():.1f}k   Clean median: {np.median(clean):.1f}k")

# Line chart
months = np.arange(1, 25)
plt.figure(figsize=(11, 5))
plt.plot(months, views, marker="o", color="steelblue", label="Monthly Views")
plt.plot(months[outlier_idx], views[outlier_idx], "ro", markersize=10, label="Outlier")
plt.axhline(clean.mean(), color="green", linestyle="--",
            label=f"Clean Mean = {clean.mean():.0f}k")
plt.title("YouTube Monthly Views — Outlier Detection")
plt.xlabel("Month")
plt.ylabel("Views (thousands)")
plt.legend()
plt.tight_layout()
plt.savefig("medium_11_youtube_outlier.png", dpi=100)
plt.show()

# ── WHY ────────────────────────────────────────────────────────────────────
# Z-score = number of SDs from mean.  |z| > 2 captures ~5% of a normal dist.
# The mean is NOT resistant to outliers because it sums every value.
# The median only depends on the middle value(s) after sorting — one extreme
# value cannot significantly shift it.
```
