Quiz 28 — Normal Distribution  [SOLVED]
Theme  : University Grades

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(21)

# normal(loc=mean, scale=std, size)
grades = np.random.normal(loc=65, scale=12, size=200)
grades = np.clip(grades, 0, 100)       # enforce valid range

print(f"Actual mean: {grades.mean():.2f}  |  Actual std: {grades.std():.2f}")
print("Distinctions (>=80):", (grades >= 80).sum())
print("Fails (<40)         :", (grades < 40).sum())

# Visualisation
plt.figure(figsize=(9, 5))
plt.hist(grades, bins=15, color="steelblue", edgecolor="white", alpha=0.8)
plt.axvline(grades.mean(), color="red", linewidth=2, label=f"Mean = {grades.mean():.1f}")
plt.axvline(grades.mean() - grades.std(), color="orange", linestyle="--", label="-1 SD")
plt.axvline(grades.mean() + grades.std(), color="orange", linestyle="--", label="+1 SD")
plt.title("Distribution of Exam Grades (N=200)")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.savefig("medium_08_grades_hist.png", dpi=100)
plt.show()

# ── WHY ────────────────────────────────────────────────────────────────────
# np.random.normal generates samples from a Gaussian bell curve.
# np.clip replaces values < 0 with 0 and > 100 with 100 in one pass.
# With 200 samples, observed mean/std won't be exactly 65/12 due to sampling
# variation — this is normal and shrinks with larger N.
```
