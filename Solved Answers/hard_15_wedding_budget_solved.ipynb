```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(180)
n        = 120
table_no = np.random.randint(1, 13, n)
menu     = np.random.choice(["Meat","Fish","Veg"], n, p=[0.5,0.3,0.2])
gift_amt = np.random.normal(120, 60, n).clip(0, 500)
age      = np.random.randint(5, 80, n)

# ── 1. Menu cost ──────────────────────────────────────────────────────────────
cost_map  = {"Meat": 65, "Fish": 70, "Veg": 55}
menu_cost = np.select([menu=="Meat", menu=="Fish"], [65, 70], default=55).astype(float)
print(f"Total menu cost: £{menu_cost.sum():,.2f}")

# ── 2. Net cost ───────────────────────────────────────────────────────────────
net = menu_cost - gift_amt
print(f"Total net cost : £{net.sum():,.2f}  "
      f"({'couple pays' if net.sum()>0 else 'gifts exceed costs'})")

# ── 3. Table-level analysis ───────────────────────────────────────────────────
print("\nTable analysis:")
best_table_net = np.inf
best_table     = -1
for t in range(1, 13):
    m    = table_no == t
    gift = gift_amt[m].sum()
    cost = menu_cost[m].sum()
    nt   = cost - gift
    if nt < best_table_net:
        best_table_net = nt
        best_table     = t
    print(f"  T{t:2d}: guests={m.sum():2d}  gifts=£{gift:,.0f}  "
          f"cost=£{cost:,.0f}  net=£{nt:,.0f}")
print(f"Most generous table: Table {best_table} (net £{best_table_net:,.0f})")

# ── 4. Age segmentation ───────────────────────────────────────────────────────
seg_labels = np.select([age<18, age<65], ["Children","Adults"], default="Seniors")
print("\nAge group analysis:")
for seg in ["Children","Adults","Seniors"]:
    m = seg_labels == seg
    if m.sum() == 0:
        continue
    choices, counts = np.unique(menu[m], return_counts=True)
    pref = choices[np.argmax(counts)]
    print(f"  {seg:9s}: n={m.sum():2d}  mean_gift=£{gift_amt[m].mean():.0f}  "
          f"pref={pref}")

# ── 5. Stacked bar: menu choices per table ────────────────────────────────────
tables    = range(1, 13)
meat_c    = np.array([(( table_no==t) & (menu=="Meat")).sum() for t in tables])
fish_c    = np.array([((table_no==t) & (menu=="Fish")).sum() for t in tables])
veg_c     = np.array([((table_no==t) & (menu=="Veg")).sum()  for t in tables])

x = np.arange(1, 13)
plt.figure(figsize=(12, 5))
plt.bar(x, meat_c, label="Meat",  color="tomato")
plt.bar(x, fish_c, bottom=meat_c, label="Fish",  color="steelblue")
plt.bar(x, veg_c,  bottom=meat_c+fish_c, label="Veg", color="green")
plt.xticks(x, [f"T{t}" for t in x])
plt.title("Wedding Menu Choices by Table")
plt.ylabel("Guests")
plt.legend()
plt.tight_layout()
plt.savefig("hard_15_wedding_menu.png", dpi=100)
plt.show()
```
