Quiz 01 — NumPy Arrays + Indexing  [SOLVED]
Theme  : Call Center
Topics : np.array, basic indexing, np.max / np.sum

```python
import numpy as np

calls = np.array([34, 45, 23, 67, 89, 56, 45, 34, 78, 90, 12, 55])

# 1. np.sum adds every element in the array
print("Total calls :", np.sum(calls))          # 628

# 2. Index 0 = first hour, index -1 = last (Python negative indexing)
print("First hour  :", calls[0], " | Last hour :", calls[-1])   # 34 | 55

# 3. Slice [3:8] gives indices 3,4,5,6,7 → hours 4-8 (1-indexed)
print("Hours 4-8   :", calls[3:8])             # [67 89 56 45 34]

# 4. np.max scans the whole array and returns the single largest value
print("Busiest hr  :", np.max(calls))          # 90

# ── WHY ────────────────────────────────────────────────────────────────────
# np.sum / np.max operate on every element without a Python loop.
# Slice notation [start:stop] is exclusive of stop, so [3:8] gives 5 values.
# Negative index -1 always refers to the last element regardless of length.
```
