Quiz 12 — Filtering Arrays  [SOLVED]
Theme  : Football Player Stats

```python
import numpy as np

goals = np.array([5, 12, 3, 18, 7, 22, 9, 15, 1, 11])
ages  = np.array([24, 28, 22, 31, 26, 29, 23, 27, 20, 25])

# Single condition
print("Over 10 goals       :", goals[goals > 10])

# Compound AND — each condition must be wrapped in parentheses
young_productive = goals[(ages < 27) & (goals > 5)]
print("Young + productive  :", young_productive)

# Compound OR
extremes = goals[(goals < 5) | (goals > 15)]
print("Few or many goals   :", extremes)

# ── WHY ────────────────────────────────────────────────────────────────────
# & is element-wise AND (not Python's 'and'); | is element-wise OR.
# Parentheses are required because & binds tighter than comparison operators.
# Both conditions are evaluated as full boolean arrays, then combined.
```
