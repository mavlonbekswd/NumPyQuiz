Quiz 11 — Boolean Masking  [SOLVED]
Theme  : Gym Members

```python
import numpy as np

ages = np.array([17, 23, 35, 42, 19, 28, 55, 31, 16, 45, 26, 38])

# Comparison operators on NumPy arrays return a boolean array element-wise
mask = ages >= 18
print("Mask   :", mask)

# Passing a boolean array as index selects only where mask is True
adults = ages[mask]
print("Adults :", adults)

# True counts as 1, False as 0 — summing the mask = count of True values
print("Count  :", mask.sum())   # or np.count_nonzero(mask)

# ── WHY ────────────────────────────────────────────────────────────────────
# Boolean masking is the NumPy equivalent of SQL WHERE.
# ages >= 18 is evaluated for every element simultaneously (vectorised).
# ages[mask] returns only the elements where mask is True — no loop needed.
```
