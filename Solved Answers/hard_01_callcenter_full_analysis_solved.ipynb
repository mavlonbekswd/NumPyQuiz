```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.randint(2, 25, 20 * 8 * 6).reshape(20, 8, 6)

# ── 1. Daily totals ─────────────────────────────────────────────────────────
daily_totals = data.sum(axis=(1, 2))             # collapse hours + agents
top3_idx     = np.argsort(daily_totals)[-3:][::-1]
print("Top-3 busiest days (index):", top3_idx)
print("Their totals:", daily_totals[top3_idx])

# ── 2. Hourly averages (across all days and agents) ─────────────────────────
hourly_avg  = data.mean(axis=(0, 2))             # shape (8,)
busiest_hr  = np.argmax(hourly_avg)
print(f"\nBusiest hour-slot: {busiest_hr}  (avg {hourly_avg[busiest_hr]:.1f} calls)")

# ── 3. Agent rankings ────────────────────────────────────────────────────────
agent_totals = data.sum(axis=(0, 1))             # shape (6,)
rank_order   = np.argsort(agent_totals)[::-1]    # descending
print("\nAgent ranking (best first):")
for rank, agent in enumerate(rank_order):
    print(f"  Rank {rank+1}: Agent {agent+1}  Total={agent_totals[agent]}")

# ── 4. Peak hours (team total > 100 in one hour-slot on one day) ─────────────
hourly_team  = data.sum(axis=2)                  # shape (20,8) — team per slot
peak_days, peak_hrs = np.where(hourly_team > 100)
print(f"\nPeak hours (team > 100 calls): {len(peak_days)} occurrences")
for d, h in zip(peak_days[:5], peak_hrs[:5]):    # show first 5
    print(f"  Day {d+1}, Hour-slot {h+1}: {hourly_team[d,h]} calls")

# ── 5. Percentiles of daily totals ───────────────────────────────────────────
q25, q50, q75 = np.percentile(daily_totals, [25, 50, 75])
print(f"\nDaily totals — Q1: {q25:.0f}  Median: {q50:.0f}  Q3: {q75:.0f}")

# ── 6. Bar chart ─────────────────────────────────────────────────────────────
colours = ["green" if t >= q50 else "tomato" for t in daily_totals]
days    = np.arange(1, 21)

plt.figure(figsize=(12, 5))
plt.bar(days, daily_totals, color=colours, edgecolor="white")
plt.axhline(q50, color="navy", linestyle="--", linewidth=2,
            label=f"Median = {q50:.0f}")
plt.title("Call Centre — Daily Total Calls (20 Working Days)")
plt.xlabel("Day")
plt.ylabel("Total Calls")
plt.legend()
plt.tight_layout()
plt.savefig("hard_01_callcenter_daily.png", dpi=100)
plt.show()

# ── WHY ───────────────────────────────────────────────────────────────────────
# axis=(1,2) collapses two axes simultaneously — equivalent to chaining two sums.
# np.where on a 2-D array returns two index arrays (rows, cols) simultaneously.
# argsort()[-3:][::-1] gets the last 3 of ascending sort, then reverses for top-3.
```
