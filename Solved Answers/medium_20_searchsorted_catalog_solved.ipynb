Quiz 40 — searchsorted + Tiered Pricing  [SOLVED]
Theme  : E-commerce Product Catalog

```python
import numpy as np

tier_bounds  = np.array([0, 25, 50, 100, 200])
tier_labels  = ["£4.99","£3.99","£2.99","£0.99","FREE"]
tier_costs   = np.array([4.99, 3.99, 2.99, 0.99, 0.00])
orders       = np.array([18.5, 55.0, 120.0, 9.99, 200.0, 75.0, 45.0, 250.0, 30.0, 15.0])

# searchsorted returns insertion points in a sorted array
# 'right' side: 200.0 maps to index 5, but we cap at 4 (max tier index)
indices = np.searchsorted(tier_bounds, orders, side="right") - 1
# Clip to valid range [0, len(tier_labels)-1]
indices = np.clip(indices, 0, len(tier_labels) - 1)

print("Order → Shipping Cost:")
for order, idx in zip(orders, indices):
    print(f"  £{order:.2f}  → {tier_labels[idx]}")

free_count = (orders >= 200).sum()
print(f"\nFree shipping orders: {free_count}")

shipping_revenue = tier_costs[indices].sum()
print(f"Total shipping revenue: £{shipping_revenue:.2f}")

# ── WHY ────────────────────────────────────────────────────────────────────
# searchsorted(bins, values, side='right') returns how far right each value
# would be inserted to keep the bins sorted.
# Subtracting 1 converts insertion point → bracket index.
# np.clip handles edge cases (value below lowest bin → 0, above highest → max).
```
