```python
import numpy as np

np.random.seed(170)
n            = 500
distance_km  = np.random.uniform(1, 40, n)
time_min     = distance_km * np.random.uniform(2.5, 4.5, n)
tip          = np.random.choice([0,1,2,3,5], n, p=[0.30,0.25,0.20,0.15,0.10])
surge        = np.random.choice([1.0,1.5,2.0], n, p=[0.60,0.30,0.10])

# ── 1. Fare computation ───────────────────────────────────────────────────────
fare       = 2.50 + 1.80*distance_km + 0.25*time_min + tip
final_fare = fare * surge

# ── 2. Summary stats ──────────────────────────────────────────────────────────
print(f"Total revenue : £{final_fare.sum():,.2f}")
print(f"Mean fare     : £{final_fare.mean():.2f}")
print(f"Median fare   : £{np.median(final_fare):.2f}")
print(f"Std           : £{final_fare.std():.2f}")

# ── 3. Surge tier analysis ────────────────────────────────────────────────────
print("\nSurge tier breakdown:")
for s in [1.0, 1.5, 2.0]:
    m = surge == s
    print(f"  {s}x: {m.sum():3d} trips  mean fare=£{final_fare[m].mean():.2f}")

# ── 4. Distance segments ──────────────────────────────────────────────────────
segs = np.select([distance_km < 5, distance_km <= 15], ["Short","Medium"], default="Long")
print("\nDistance segment breakdown:")
for seg in ["Short","Medium","Long"]:
    m = segs == seg
    print(f"  {seg:6s}: {m.sum():3d} trips  "
          f"mean_fare=£{final_fare[m].mean():.2f}  mean_tip=£{tip[m].mean():.2f}")

# ── 5. High-fare trips ────────────────────────────────────────────────────────
p90       = np.percentile(final_fare, 90)
high_mask = final_fare > p90
surge_pct = (surge[high_mask] > 1.0).mean() * 100
print(f"\nFare > P90 (£{p90:.2f}): {high_mask.sum()} trips  "
      f"with surge: {surge_pct:.1f}%")

# ── 6. Profit margin ─────────────────────────────────────────────────────────
driver_cost = 0.35*distance_km + 0.10*time_min
profit      = final_fare - driver_cost
margin_pct  = profit.sum() / final_fare.sum() * 100
print(f"\nTotal profit  : £{profit.sum():,.2f}")
print(f"Profit margin : {margin_pct:.1f}%")
```
