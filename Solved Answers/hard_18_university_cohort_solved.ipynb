Quiz 58 — University Multi-Year Cohort Comparison  [SOLVED]
Difficulty: Hard

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(210)
loc    = np.array([[60,65,58,70],[62,68,60,72],[64,70,62,74]])
scores = np.random.normal(loc[:,:,np.newaxis], 12, (3,4,50)).clip(0, 100)
subjects = ["Maths","Physics","CS","Stats"]

# ── 1. Year-on-year improvement (mean per subject) ────────────────────────────
means_ys = scores.mean(axis=2)              # (3,4) — mean per year+subject
yoy21    = means_ys[1] - means_ys[0]        # yr2 - yr1
yoy32    = means_ys[2] - means_ys[1]        # yr3 - yr2
print("Year-on-year mean improvement:")
for i, sub in enumerate(subjects):
    print(f"  {sub:8s}: Yr2-Yr1={yoy21[i]:+.2f}  Yr3-Yr2={yoy32[i]:+.2f}")

# ── 2. Distinction rate per subject per year ──────────────────────────────────
print("\nDistinction rate (>=70):")
for yr in range(3):
    rates = (scores[yr] >= 70).mean(axis=1) * 100   # per subject
    print(f"  Year {yr+1}: " + "  ".join(f"{s}:{r:.0f}%" for s,r in zip(subjects,rates)))

# ── 3. Best/worst subject ─────────────────────────────────────────────────────
overall_means = scores.mean(axis=(0,2))     # mean over years and students → (4,)
best_sub  = np.argmax(overall_means)
worst_sub = np.argmin(overall_means)
print(f"\nBest  subject: {subjects[best_sub]}  ({overall_means[best_sub]:.1f})")
print(f"Worst subject: {subjects[worst_sub]} ({overall_means[worst_sub]:.1f})")

# ── 4. Students who improved every year in Maths ─────────────────────────────
maths = scores[:, 0, :]                     # (3, 50) — maths scores per year
improved = (maths[1] > maths[0]) & (maths[2] > maths[1])
print(f"\nStudents improving in Maths both years: {improved.sum()}")
print("Their indices:", np.where(improved)[0][:10], "...")

# ── 5. Grouped bar chart ──────────────────────────────────────────────────────
x     = np.arange(len(subjects))
width = 0.25
colours = ["steelblue","orange","green"]

fig, ax = plt.subplots(figsize=(10, 6))
for yr in range(3):
    ax.bar(x + (yr-1)*width, means_ys[yr], width,
           label=f"Year {yr+1}", color=colours[yr], alpha=0.85)
ax.set_xticks(x)
ax.set_xticklabels(subjects)
ax.set_ylabel("Mean Score")
ax.set_title("University Score by Subject and Year")
ax.legend()
plt.tight_layout()
plt.savefig("hard_18_university_cohort.png", dpi=100)
plt.show()
```
