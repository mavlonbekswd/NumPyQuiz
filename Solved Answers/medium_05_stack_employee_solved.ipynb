Quiz 25 — Stacking + Joining + Column Operations  [SOLVED]
Theme  : Employee Salaries

```python
import numpy as np

emp_ids  = np.array([101,102,103,104,105,106,107,108])
salaries = np.array([3200,4500,2800,5100,3800,4200,3600,4900])
bonuses  = np.array([320, 450, 280, 510, 380, 420, 360, 490])

total_pay = salaries + bonuses

# column_stack turns 1-D arrays into columns of a 2-D matrix
table = np.column_stack([emp_ids, salaries, bonuses, total_pay])
print("Initial shape:", table.shape)   # (8, 4)

# Add three more employees
new_data = np.array([[109,3000,300,3300],
                     [110,5500,550,6050],
                     [111,2500,250,2750]])
table = np.vstack([table, new_data])
print("Final shape:", table.shape)     # (11, 4)

# column 3 is total_pay; argmax → row index; column 0 → emp_id
best_row = np.argmax(table[:, 3])
print(f"Highest paid ID: {int(table[best_row, 0])}  Total: {int(table[best_row, 3])}")

# ── WHY ────────────────────────────────────────────────────────────────────
# np.column_stack is the cleanest way to build a table from separate arrays.
# vstack appends rows — the column count must match (4 in both cases here).
# table[:, 3] selects all rows of column 3 (total_pay).
```
