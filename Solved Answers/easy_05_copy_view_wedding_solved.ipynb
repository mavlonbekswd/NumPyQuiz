Quiz 05 — Copy vs View  [SOLVED]
Theme  : Wedding Guests
Topics : view (slice), copy(), .base attribute

```python
import numpy as np

ages = np.array([22, 35, 28, 45, 30, 52, 18, 61, 40, 33])

# A slice returns a VIEW — it shares memory with the original
view = ages[:5]

# .copy() allocates new memory — completely independent
copy = ages[5:].copy()

# Modify view → original is affected
view[0] = 99
print("After view change  :", ages)   # [99 35 28 45 30 52 18 61 40 33]

# Modify copy → original is NOT affected
copy[0] = 99
print("After copy change  :", ages)   # still [99 35 28 45 30 52 18 61 40 33]

# .base is not None when the array shares data with another object
print("view.base is ages  :", view.base is ages)   # True
print("copy.base is None  :", copy.base is None)   # True

# ── WHY ────────────────────────────────────────────────────────────────────
# Views avoid memory duplication — efficient for large arrays, but
# accidental modification of the original is a common bug.
# Always use .copy() when you need an independent dataset to modify safely.
```
