Quiz 57 — E-commerce Return Rate Analysis  [SOLVED]
Difficulty: Hard

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(200)
categories   = ["Electronics","Clothing","Books","Home","Toys"]
return_rates = np.array([0.12, 0.25, 0.05, 0.08, 0.15])
orders       = np.random.randint(200, 1000, (5, 6))

# Generate binomial returns for each cell
returns = np.zeros((5, 6), dtype=int)
for i in range(5):
    for j in range(6):
        returns[i, j] = np.random.binomial(orders[i, j], return_rates[i])

# ── 1. Actual return rates (5,6) ──────────────────────────────────────────────
actual_rates = returns / orders              # element-wise

# ── 2. Flag cells > 2x expected rate ─────────────────────────────────────────
# expected_rates (5,) → reshape (5,1) to broadcast over (5,6)
expected     = return_rates[:, np.newaxis]
double_rate  = actual_rates > (2 * expected)
print(f"Cells exceeding 2x expected rate: {double_rate.sum()}")
for cat_i, mon_j in zip(*np.where(double_rate)):
    print(f"  {categories[cat_i]} Month {mon_j+1}: "
          f"actual={actual_rates[cat_i,mon_j]:.2%} vs expected={return_rates[cat_i]:.2%}")

# ── 3. Return processing cost ─────────────────────────────────────────────────
return_cost = returns * 15               # £15 per return
total_cost_per_cat = return_cost.sum(axis=1)
print("\nReturn processing cost per category:")
for name, cost in zip(categories, total_cost_per_cat):
    print(f"  {name:13s}: £{cost:,.0f}")

# ── 4. Audit flags (> 20% return rate) ───────────────────────────────────────
audit_flags = actual_rates > 0.20
if audit_flags.any():
    print("\n⚠ Audit flags (>20% return rate):")
    for ci, mj in zip(*np.where(audit_flags)):
        print(f"  {categories[ci]} Month {mj+1}: {actual_rates[ci,mj]:.1%}")

# ── 5. Grouped bar chart ──────────────────────════════════════════════════════
mean_actual   = actual_rates.mean(axis=1)    # mean across 6 months per category
x             = np.arange(len(categories))
width         = 0.35

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(x - width/2, return_rates * 100, width, label="Expected Rate", color="steelblue")
ax.bar(x + width/2, mean_actual * 100,  width, label="Actual Mean Rate", color="tomato")
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=15)
ax.set_ylabel("Return Rate (%)")
ax.set_title("Return Rate: Expected vs Actual (6-month avg)")
ax.legend()
plt.tight_layout()
plt.savefig("hard_17_returns_chart.png", dpi=100)
plt.show()
```
