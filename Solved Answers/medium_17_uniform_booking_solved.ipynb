```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(23)
# uniform(low, high, size) — every value equally probable between low and high
wait_uniform = np.random.uniform(1, 15, 500)
wait_normal  = np.random.normal(8, 2, 500)

for label, arr in [("Uniform", wait_uniform), ("Normal", wait_normal)]:
    print(f"{label} — Mean: {arr.mean():.2f}  Std: {arr.std():.2f}")

print(f"\nUniform > 10 min: {(wait_uniform > 10).sum()}")
print(f"Normal  > 10 min: {(wait_normal  > 10).sum()}")

plt.figure(figsize=(10, 5))
plt.hist(wait_uniform, bins=25, alpha=0.6, color="steelblue", label="Uniform(1,15)")
plt.hist(wait_normal,  bins=25, alpha=0.6, color="tomato",    label="Normal(8,2)")
plt.axvline(wait_uniform.mean(), color="blue",  linestyle="--",
            label=f"Uniform mean={wait_uniform.mean():.1f}")
plt.axvline(wait_normal.mean(),  color="red",   linestyle="--",
            label=f"Normal mean={wait_normal.mean():.1f}")
plt.title("Booking Wait Time Distributions (n=500)")
plt.xlabel("Minutes")
plt.ylabel("Count")
plt.legend()
plt.tight_layout()
plt.savefig("medium_17_uniform_vs_normal.png", dpi=100)
plt.show()

# ── WHY ────────────────────────────────────────────────────────────────────
# Uniform: std = (high-low)/√12 ≈ (15-1)/3.46 ≈ 4.04 — flat histogram.
# Normal: std=2, concentrated around the mean — bell-shaped histogram.
# Both have similar means but very different spreads and shapes.
```
