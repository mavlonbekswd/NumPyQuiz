Quiz 24 — Reshape + Axis Operations  [SOLVED]
Theme  : Call Centre Shifts

```python
import numpy as np

np.random.seed(3)
raw  = np.random.randint(5, 30, 60)
data = raw.reshape(5, 2, 3)          # (days, shifts, agents)
print("Shape:", data.shape)

# axis=(0,1) collapses both days and shifts, leaving one mean per agent
agent_means = data.mean(axis=(0, 1))
print("Mean per agent (all days+shifts):", np.round(agent_means, 2))

# sum over shifts (axis=1) then over agents (axis=-1)
day_totals = data.sum(axis=(1, 2))   # collapse shifts and agents
print("Total calls per day:", day_totals)

busiest = np.argmax(day_totals)
print("Busiest day index:", busiest)

# ── WHY ────────────────────────────────────────────────────────────────────
# axis=(0,1) passes a tuple — NumPy collapses both axes at once.
# Alternatively: data.sum(axis=1).sum(axis=1) gives the same result.
# 3-D arrays are common with time-series data (e.g. weeks × days × hours).
```
