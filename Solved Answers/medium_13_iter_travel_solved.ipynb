Quiz 33 — nditer + Structured Iteration  [SOLVED]
Theme  : Travel Bookings

```python
import numpy as np

prices = np.array([[499, 599, 749, 399],
                   [899, 999,1199, 799],
                   [1499,1799,2199,1299]], dtype=float)

packages = ["Budget","Standard","Premium"]
seasons  = ["Winter","Spring","Summer","Autumn"]

# readwrite flag allows modifying elements through the iterator
with np.nditer(prices, op_flags=["readwrite"]) as it:
    for x in it:
        x[...] *= 1.05   # x[...] modifies in-place; x alone would not

print("Prices after 5% increase:")
print(np.round(prices, 2))

# ndenumerate gives (row, col) index + value
print("\nFull price list:")
for (pkg_i, sea_i), price in np.ndenumerate(prices):
    print(f"  {packages[pkg_i]} / {seasons[sea_i]}: £{price:.0f}")

# Multi-array nditer — iterate two arrays element-by-element
tax = prices * 0.10
print("\nPrice + Tax:")
for p, t in np.nditer([prices, tax]):
    print(f"  Price: £{float(p):.0f}  Tax: £{float(t):.0f}")

# Highest price
flat_idx       = np.argmax(prices)
pkg_i, sea_i   = np.unravel_index(flat_idx, prices.shape)
print(f"\nMost expensive: {packages[pkg_i]} / {seasons[sea_i]}: £{prices[pkg_i, sea_i]:.0f}")

# ── WHY ────────────────────────────────────────────────────────────────────
# x[...] (ellipsis assignment) writes back through the iterator buffer.
# Without [...] you'd just rebind the local variable x, leaving prices unchanged.
```
