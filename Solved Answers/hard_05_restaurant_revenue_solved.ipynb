```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(80)
covers      = np.random.randint(0, 20, 4*7*3).reshape(4, 7, 3)
session_avg = np.array([12, 22, 38])

# ── 1. Revenue via broadcasting ───────────────────────────────────────────────
# covers: (4,7,3)  session_avg: (3,) → broadcast as (1,1,3)
revenue = covers * session_avg              # (4,7,3)
print(f"Total monthly revenue: £{revenue.sum():,.0f}")

# ── 2. Weekly totals ──────────────────────────────────────────────────────────
weekly = revenue.sum(axis=(1, 2))           # collapse days + sessions
print("Weekly revenue:", np.round(weekly, 0))
best_week = np.argmax(weekly)
print(f"Best week: Week {best_week+1} (£{weekly[best_week]:,.0f})")

# ── 3. Best day of week ───────────────────────────────────────────────────────
day_names  = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
daily_rev  = revenue.sum(axis=2)            # (4,7) — per day revenue
mean_by_dow = daily_rev.mean(axis=0)        # average across 4 weeks → (7,)
best_day    = np.argmax(mean_by_dow)
print(f"Best day of week: {day_names[best_day]} (avg £{mean_by_dow[best_day]:.0f})")

# ── 4. Best session ───────────────────────────────────────────────────────────
session_names = ["Breakfast","Lunch","Dinner"]
session_rev   = revenue.sum(axis=(0,1))     # (3,)
best_sess     = np.argmax(session_rev)
print(f"Best session: {session_names[best_sess]} (£{session_rev[best_sess]:,.0f})")

# ── 5. Closure days ───────────────────────────────────────────────────────────
# A day has zero covers in all sessions → row sum == 0
day_covers = covers.sum(axis=2)             # (4,7) — total covers per day
closure    = np.where(day_covers == 0)      # returns (week_idx_arr, day_idx_arr)
print(f"Closure days: {len(closure[0])}")
for wk, dy in zip(*closure):
    print(f"  Week {wk+1}, {day_names[dy]}")

# ── 6. Heatmap ────────────────────────────────────────────────────────────────
plt.figure(figsize=(10, 4))
plt.imshow(daily_rev, aspect="auto", cmap="YlOrRd")
plt.colorbar(label="Daily Revenue (£)")
plt.xticks(range(7), day_names)
plt.yticks(range(4), [f"Week {i+1}" for i in range(4)])
plt.title("Restaurant Daily Revenue Heatmap (4 Weeks)")
plt.tight_layout()
plt.savefig("hard_05_restaurant_heatmap.png", dpi=100)
plt.show()

# ── WHY ───────────────────────────────────────────────────────────────────────
# session_avg (3,) broadcasts to (4,7,3) because NumPy aligns from the right:
# last axis of covers (size 3) matches session_avg (size 3).
# revenue.sum(axis=(0,1)) collapses weeks and days simultaneously.
```
