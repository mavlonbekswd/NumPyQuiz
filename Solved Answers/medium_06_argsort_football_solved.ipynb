Quiz 26 — argsort + Fancy Indexing  [SOLVED]
Theme  : Football Player Stats

```python
import numpy as np

players = np.array(["Tomas","Ivan","Luca","Kai","Seb","Jorge","Nico","Ali","Finn","Zak"])
goals   = np.array([15, 22, 8, 18, 12, 25, 9, 30, 14, 20])
assists = np.array([ 7, 10,12,  8, 15,  6,11,  5,  9, 13])

# argsort returns ascending; [::-1] reverses for descending
goal_rank = np.argsort(goals)[::-1]
print("Goals ranking:")
for name, g in zip(players[goal_rank], goals[goal_rank]):
    print(f"  {name}: {g}")

# Contribution score
contrib = goals + assists
contrib_rank = np.argsort(contrib)[::-1]
print("\nTop 5 by contribution:")
for name, c in zip(players[contrib_rank[:5]], contrib[contrib_rank[:5]]):
    print(f"  {name}: {c}")

# lexsort: LAST key in the tuple is the PRIMARY sort key
# negate for descending
lex_order = np.lexsort((-goals, -assists))   # primary: assists desc, secondary: goals desc
print("\nLexsort (assists desc, then goals desc):")
for name, a, g in zip(players[lex_order], assists[lex_order], goals[lex_order]):
    print(f"  {name}: assists={a}, goals={g}")

# ── WHY ────────────────────────────────────────────────────────────────────
# argsort gives positions — fancy-indexing a string array with those positions
# reorders names to match the sorted numeric order.
# np.lexsort takes a tuple of keys; the LAST key is applied first (primary).
```
