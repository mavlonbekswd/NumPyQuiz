Quiz 30 — Percentile + Quartiles  [SOLVED]
Theme  : Employee Salaries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(11)
salaries = np.random.randint(22000, 95000, 25)

q1  = np.percentile(salaries, 25)
q2  = np.percentile(salaries, 50)   # same as np.median
q3  = np.percentile(salaries, 75)
p90 = np.percentile(salaries, 90)
print(f"Q1: £{q1:.0f}   Q2: £{q2:.0f}   Q3: £{q3:.0f}   P90: £{p90:.0f}")

iqr = q3 - q1
print(f"IQR: £{iqr:.0f}")

lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = salaries[(salaries < lower) | (salaries > upper)]
if len(outliers):
    print("Outliers:", outliers)
else:
    print("No outliers detected")

# Boxplot shows Q1, Q2, Q3, whiskers (1.5*IQR), and fliers
plt.figure(figsize=(7, 5))
sns.boxplot(x=salaries, color="lightblue")
plt.axvline(q1,  color="green",  linestyle="--", label=f"Q1 £{q1:.0f}")
plt.axvline(q2,  color="blue",   linestyle="-",  label=f"Q2 £{q2:.0f}")
plt.axvline(q3,  color="purple", linestyle="--", label=f"Q3 £{q3:.0f}")
plt.title("Employee Salary Distribution — Quartile Analysis")
plt.xlabel("Annual Salary (£)")
plt.legend()
plt.tight_layout()
plt.savefig("medium_10_salary_boxplot.png", dpi=100)
plt.show()

# ── WHY ────────────────────────────────────────────────────────────────────
# np.percentile uses linear interpolation by default — the 50th percentile
# equals np.median for the same data.
# The 1.5*IQR rule (Tukey's fence) is the standard definition of mild outliers
# used by most boxplot implementations.
```
