Quiz 19 — Iteration + nditer  [SOLVED]
Theme  : Restaurant Menu

```python
import numpy as np

menu = np.array([[8.5,  12.0,  9.5, 11.0],
                 [18.0, 22.5, 16.0, 24.0],
                 [6.5,   7.0,  8.0,  9.5]])

course_names = ["Starters", "Mains", "Desserts"]

# Iterating over a 2-D array gives one row at a time
print("--- Courses ---")
for name, row in zip(course_names, menu):
    print(f"{name}: {row}")

# nditer flattens to a single stream of scalars — good for element-wise work
print("\n--- Prices with VAT ---")
for price in np.nditer(menu):
    vat = round(float(price) * 0.20, 2)
    print(f"  Price: £{float(price):.2f}  VAT: £{vat}")

# ndenumerate yields ((row, col), value) tuples
print("\n--- Index + Price ---")
for (r, c), price in np.ndenumerate(menu):
    print(f"  [{r},{c}] = £{price:.2f}")

print("\nTotal revenue:", np.sum(menu))

# ── WHY ────────────────────────────────────────────────────────────────────
# Prefer vectorised operations over iteration for performance.
# nditer / ndenumerate are useful when you need the index or complex per-element
# logic that can't easily be vectorised.
```
