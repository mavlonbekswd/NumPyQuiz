Difficulty: Hard

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(130)
n           = 300
book_value  = np.random.uniform(200, 3000, n)
nights      = np.random.randint(2, 15, n)
destination = np.random.choice(["Europe","Asia","Americas","Africa"], n)
cancelled   = np.random.choice([0, 1], n, p=[0.88, 0.12])

# ── 1. Revenue ────────────────────────────────────────────────────────────────
revenue = book_value * (1 - cancelled)               # cancelled → 0 revenue
print(f"Total revenue: £{revenue.sum():,.2f}  "
      f"(from {(cancelled == 0).sum()} active bookings)")

# ── 2. Per-destination stats ─────────────────────────────────────────────────
print("\nDestination breakdown:")
for dest in ["Europe","Asia","Americas","Africa"]:
    mask    = destination == dest
    n_dest  = mask.sum()
    mean_v  = book_value[mask].mean()
    canc_r  = cancelled[mask].mean() * 100
    print(f"  {dest:10s}: {n_dest:3d} bookings  "
          f"mean=£{mean_v:.0f}  cancel={canc_r:.1f}%")

# ── 3. Budget bookings cancellation ──────────────────────────────────────────
p10     = np.percentile(book_value, 10)
budget  = book_value < p10
print(f"\nBudget threshold (P10): £{p10:.2f}")
print(f"Budget bookings: {budget.sum()}  Cancelled: {(budget & (cancelled==1)).sum()}")

# ── 4. Value per night (active only) ─────────────────────────────────────────
active_mask = cancelled == 0
vpn         = book_value[active_mask] / nights[active_mask]
top5_vpn    = np.argsort(vpn)[-5:][::-1]
active_idx  = np.where(active_mask)[0]                # original indices
print("\nTop-5 value-per-night bookings:")
for i in top5_vpn:
    orig = active_idx[i]
    print(f"  Booking {orig+1:3d}: £{book_value[orig]:.0f} / "
          f"{nights[orig]} nights = £{vpn[i]:.2f}/night")

# ── 5. Seaborn boxplot ────────────────────────────────────────────────────────
import pandas as pd
df = pd.DataFrame({"Destination": destination, "BookingValue": book_value})
plt.figure(figsize=(9, 5))
sns.boxplot(data=df, x="Destination", y="BookingValue", palette="Set2")
plt.title("Booking Value by Destination")
plt.ylabel("Booking Value (£)")
plt.tight_layout()
plt.savefig("hard_10_travel_boxplot.png", dpi=100)
plt.show()

# ── WHY ───────────────────────────────────────────────────────────────────────
# revenue = book_value * (1 - cancelled) is an elegant vectorised approach:
# cancelled=0 → multiply by 1 (keep value); cancelled=1 → multiply by 0.
# vpn is computed only on the active subset using a boolean mask index.
```
