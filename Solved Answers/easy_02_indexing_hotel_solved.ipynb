Quiz 02 — Indexing + Slicing (2-D)  [SOLVED]
Theme  : Hotel Guests
Topics : 2-D array creation, row/column slicing, shape

```python
import numpy as np

guests = np.array([[2,0,1,3,2],
                   [1,1,0,2,3],
                   [3,2,1,0,1],
                   [0,1,2,3,2]])

# 1. .shape returns a tuple (rows, cols)
print("Shape        :", guests.shape)          # (4, 5)

# 2. Row index 2 = floor 3 (0-indexed)
print("Floor 3      :", guests[2])             # [3 2 1 0 1]

# 3. All rows, column index 3 → guests[:, 3]
print("Room 4 all   :", guests[:, 3])          # [3 2 0 3]

# 4. Row slice 1:3 gives rows 1 & 2; col slice 1:4 gives cols 1,2,3
print("Sub-grid     :")
print(guests[1:3, 1:4])
# [[1 0 2]
#  [2 1 0]]

# ── WHY ────────────────────────────────────────────────────────────────────
# arr[row, col] — both axes can be slices or single indices simultaneously.
# : alone means "all items along that axis."
# Slices are always exclusive of the stop index.
```
