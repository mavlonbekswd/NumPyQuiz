Quiz 38 — Binomial Distribution  [SOLVED]
Theme  : Call Centre

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(31)
n, p   = 20, 0.7
hours  = np.random.binomial(n=n, p=p, size=200)

theoretical_mean = n * p
theoretical_std  = np.sqrt(n * p * (1 - p))
print(f"Theoretical mean: {theoretical_mean}  |  Simulated mean: {hours.mean():.2f}")
print(f"Theoretical std : {theoretical_std:.3f}  |  Simulated std : {hours.std():.3f}")

prob_15_plus = (hours >= 15).sum() / len(hours)
print(f"P(resolved >= 15): {prob_15_plus:.3f}  ({(hours>=15).sum()} / 200 hours)")

plt.figure(figsize=(9, 5))
unique, counts = np.unique(hours, return_counts=True)
plt.bar(unique, counts, color="steelblue", edgecolor="white")
plt.axvline(theoretical_mean, color="red", linestyle="--",
            label=f"Expected = {theoretical_mean}")
plt.title("Call Centre: Hourly Resolved Calls (Binomial n=20, p=0.7)")
plt.xlabel("Calls Resolved")
plt.ylabel("Frequency (out of 200 hours)")
plt.legend()
plt.tight_layout()
plt.savefig("medium_18_binomial_callcenter.png", dpi=100)
plt.show()

# ── WHY ────────────────────────────────────────────────────────────────────
# Binomial(n, p) counts successes in n independent Bernoulli trials.
# Expected value = n*p = 14 (on average 14 of 20 calls resolved per hour).
# std = sqrt(n*p*(1-p)) — measures spread around the expected value.
# With 200 simulated hours the observed mean should be very close to 14.
```
