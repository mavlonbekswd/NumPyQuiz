Quiz 39 — Poisson Distribution  [SOLVED]
Theme  : Restaurant Arrivals

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(37)
lam      = 8
arrivals = np.random.poisson(lam=lam, size=100)

print(f"Theoretical mean: {lam}  |  Simulated mean: {arrivals.mean():.2f}")
print(f"Theoretical std : {np.sqrt(lam):.3f}  |  Simulated std : {arrivals.std():.3f}")

overwhelmed = (arrivals > 12).sum()
underused   = (arrivals < 5).sum()
print(f"Hours > 12 arrivals (overwhelmed): {overwhelmed}%")
print(f"Hours < 5  arrivals (underused)  : {underused}%")

unique, counts = np.unique(arrivals, return_counts=True)
plt.figure(figsize=(10, 5))
plt.bar(unique, counts, color="coral", edgecolor="white")
plt.axvline(lam, color="blue", linestyle="--", label=f"λ = {lam}")
plt.title("Restaurant Hourly Arrivals — Poisson(λ=8), n=100 hours")
plt.xlabel("Groups Arriving per Hour")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.savefig("medium_19_poisson_restaurant.png", dpi=100)
plt.show()

# ── WHY ────────────────────────────────────────────────────────────────────
# Poisson models count of independent events in a fixed interval.
# For Poisson: mean = λ AND variance = λ (unique property).
# std = sqrt(λ) ≈ 2.83 — spread grows with the rate but slower than the mean.
```
