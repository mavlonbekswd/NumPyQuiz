Quiz 10 — Mean & Median  [SOLVED]
Theme  : YouTube Analytics
Topics : np.mean, np.median, outlier sensitivity

```python
import numpy as np

views = np.array([1200, 980, 1050, 1300, 15000, 1100, 900, 1250, 1080, 970])

mean_all   = np.mean(views)
median_all = np.median(views)
print(f"With viral day  — Mean: {mean_all}  |  Median: {median_all}")

# np.delete returns a new array with the element at index 4 removed
no_viral = np.delete(views, 4)
mean_nv   = np.mean(no_viral)
median_nv = np.median(no_viral)
print(f"Without viral   — Mean: {mean_nv}  |  Median: {median_nv}")

print("Median is more representative of a typical day.")

# ── WHY ────────────────────────────────────────────────────────────────────
# Mean sums every value — one extreme outlier (15,000) pulls it far from the
# majority of days (900–1300).
# Median is the middle value after sorting; a single outlier barely moves it.
# Rule of thumb: use median when data has outliers or is skewed.
```
