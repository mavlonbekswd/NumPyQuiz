Quiz 18 — argmax + argmin  [SOLVED]
Theme  : Call Centre

```python
import numpy as np

handled = np.array([[12,15, 8,20,18,14],
                    [10,18,12,16,22,11],
                    [14,10,16,12,20,17],
                    [ 8,14,18,14,15,20]])

# Sum each row → total per agent; argmax finds which agent's total is highest
agent_totals = handled.sum(axis=1)         # [87, 89, 89, 89] — agent 1 wins on order
print("Busiest agent (index):", np.argmax(agent_totals))   # 1

# Sum each column → total per hour; argmax finds peak hour
hour_totals = handled.sum(axis=0)
print("Busiest hour  (index):", np.argmax(hour_totals))    # 3

# axis=1 → argmax operates across columns for each row independently
print("Each agent's peak hour:", np.argmax(handled, axis=1))  # [3 4 4 5]

# ── WHY ────────────────────────────────────────────────────────────────────
# argmax(axis=1) returns one index per row — the column position of the max.
# It does NOT return the max value, only its position.
# Combining sum() + argmax() is a common pattern for "which group is biggest."
```
