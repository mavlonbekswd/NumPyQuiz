Quiz 48 — Football Performance Ranking  [SOLVED]
Difficulty: Hard

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(110)
n            = 20
goals        = np.random.randint(0, 30, n)
assists      = np.random.randint(0, 20, n)
pass_acc     = np.random.uniform(60, 95, n)
tackles      = np.random.randint(10, 80, n)
rating       = np.random.uniform(5.0, 9.5, n)
player_names = np.array([f"P{i:02d}" for i in range(1, n+1)])

# ── 1. Composite score ────────────────────────────────────────────────────────
score = (goals * 3 + assists * 2 +
         (pass_acc / 100) * 10 +
         (tackles / 80)  * 5  +
         rating * 2)

# ── 2. Ranking ────────────────────────────────────────────────────────────────
rank_idx = np.argsort(score)[::-1]
print("Top-5 players:")
for i in rank_idx[:5]:
    print(f"  {player_names[i]}: score={score[i]:.2f} "
          f"goals={goals[i]} assists={assists[i]} rating={rating[i]:.1f}")

# ── 3. World Class / Average / Struggling ────────────────────────────────────
p25, p75 = np.percentile(score, [25, 75])
labels   = np.select(
    [score >= p75, score >= p25],
    ["World Class", "Average"],
    default="Struggling"
)
for lbl in ["World Class", "Average", "Struggling"]:
    count = (labels == lbl).sum()
    print(f"  {lbl:12s}: {count} players")

# ── 4. Top-half most tackles ─────────────────────────────────────────────────
med_score     = np.median(score)
top_half_mask = score >= med_score
top_tackles   = np.argmax(tackles[top_half_mask])
top_idx       = np.where(top_half_mask)[0][top_tackles]
print(f"\nMost tackles in top half: {player_names[top_idx]} ({tackles[top_idx]} tackles)")

# ── 5. Lexsort: primary=rating desc, secondary=goals desc ────────────────────
lex_idx = np.lexsort((-goals, -rating))   # last key = primary
print("\nTop-5 by Rating then Goals:")
for i in lex_idx[:5]:
    print(f"  {player_names[i]}: rating={rating[i]:.2f} goals={goals[i]}")

# ── 6. Horizontal bar chart top-10 ───────────────────────────────────────────
top10 = rank_idx[:10]
fig, ax = plt.subplots(figsize=(9, 6))
ax.barh(player_names[top10][::-1], score[top10][::-1], color="steelblue")
ax.set_xlabel("Composite Score")
ax.set_title("Top-10 Football Players — Composite Score")
plt.tight_layout()
plt.savefig("hard_08_football_ranking.png", dpi=100)
plt.show()

# ── WHY ───────────────────────────────────────────────────────────────────────
# Composite score normalises each metric to a comparable scale before weighting.
# np.lexsort last key = primary sort — counterintuitive but important to remember.
# Negating arrays before lexsort achieves descending order.
```
