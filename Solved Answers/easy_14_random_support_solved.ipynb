Quiz 14 — Random randint + Seed  [SOLVED]
Theme  : Customer Support Tickets

```python
import numpy as np

np.random.seed(42)   # ensures identical results every run

# randint(low, high, size) — high is EXCLUSIVE, so use 81 for max=80
tickets = np.random.randint(20, 81, size=7)
print("Tickets/day :", tickets)

print("Busiest     :", np.max(tickets), " | Quietest :", np.min(tickets))
print("Average     :", np.mean(tickets))
print("Weekly total:", np.sum(tickets))

# ── WHY ────────────────────────────────────────────────────────────────────
# np.random.seed(n) fixes the internal state of the random number generator
# so results are reproducible — essential for testing and sharing notebooks.
# randint high parameter is exclusive — a common off-by-one mistake.
```
