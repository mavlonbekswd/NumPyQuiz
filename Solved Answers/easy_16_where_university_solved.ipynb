Quiz 16 — np.where  [SOLVED]
Theme  : University Marks

```python
import numpy as np

scores = np.array([72, 45, 83, 38, 90, 55, 29, 61, 48, 77])

# np.where(condition, value_if_true, value_if_false)
labels = np.where(scores >= 50, "Pass", "Fail")
print("Labels  :", labels)

# Apply different additions depending on condition
bonus    = np.where(scores >= 70, 5, 0)
adjusted = scores + bonus
print("Adjusted:", adjusted)

# Counting Passes: boolean sum (True=1)
print("Passed  :", (scores >= 50).sum())

# ── WHY ────────────────────────────────────────────────────────────────────
# np.where with three arguments is a vectorised if/else — no loop needed.
# The condition produces a boolean mask; np.where picks from the two
# value arrays element by element based on that mask.
```
