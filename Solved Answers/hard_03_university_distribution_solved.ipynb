Quiz 43 — University Grade Distribution  [SOLVED]
Difficulty: Hard

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(60)
cohort_a = np.clip(np.random.normal(62, 10, 100), 0, 100)
np.random.seed(61)
cohort_b = np.clip(np.random.normal(68, 15, 100), 0, 100)

# ── 1. Summary stats ────────────────────────────────────────────────────────
for name, c in [("Cohort A", cohort_a), ("Cohort B", cohort_b)]:
    q1, q3 = np.percentile(c, [25, 75])
    print(f"{name}: Mean={c.mean():.1f} Median={np.median(c):.1f} "
          f"Std={c.std():.1f} Q1={q1:.1f} Q3={q3:.1f}")

# ── 2. Grade bands ───────────────────────────────────────────────────────────
bands       = {"Fail (<40)": lambda g: g < 40,
               "Pass 40-59": lambda g: (g >= 40) & (g < 60),
               "Merit 60-79": lambda g: (g >= 60) & (g < 80),
               "Distinction (>=80)": lambda g: g >= 80}
print()
for band, fn in bands.items():
    ca, cb = fn(cohort_a).sum(), fn(cohort_b).sum()
    print(f"  {band:22s} | A: {ca:3d}  B: {cb:3d}")

# ── 3. Cohort A above Cohort B's 75th percentile ─────────────────────────────
b_q3   = np.percentile(cohort_b, 75)
count  = (cohort_a > b_q3).sum()
print(f"\nCohort A students above Cohort B Q3 ({b_q3:.1f}): {count}")

# ── 4. Plot ──────────────────────────────────────────────────────────────────
plt.figure(figsize=(10, 6))
plt.hist(cohort_a, bins=20, alpha=0.6, color="steelblue", label="Cohort A")
plt.hist(cohort_b, bins=20, alpha=0.6, color="tomato",    label="Cohort B")
plt.axvline(cohort_a.mean(), color="blue",   linestyle="--",
            label=f"A mean={cohort_a.mean():.1f}")
plt.axvline(cohort_b.mean(), color="red",    linestyle="--",
            label=f"B mean={cohort_b.mean():.1f}")
plt.title("University Exam Grade Distributions")
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.savefig("hard_03_university_cohorts.png", dpi=100)
plt.show()

# ── 5. Conclusion ────────────────────────────────────────────────────────────
if cohort_b.mean() > cohort_a.mean():
    print("\nCohort B has a higher mean — but also higher spread (std).")
    print("B has more distinctions AND more failures; A is more consistent.")

# ── WHY ───────────────────────────────────────────────────────────────────────
# Mean alone does not tell the full story — a higher mean with higher std can
# mean more extreme outcomes in both directions.
# Q1/Q3 alongside mean/median give a richer picture of distribution shape.
```
