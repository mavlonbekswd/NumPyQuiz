Quiz 04 — Reshape  [SOLVED]
Theme  : Grocery Sales
Topics : reshape, flatten

```python
import numpy as np

sales = np.array([120, 200, 150, 170, 230, 190, 140, 160, 210, 220, 180, 250])
print("Before reshape :", sales.shape)         # (12,)

# reshape does NOT copy data — it returns a view with a new shape
# rows × cols must equal original size: 3 × 4 = 12 ✓
weekly = sales.reshape(3, 4)
print("After reshape  :", weekly.shape)        # (3, 4)

# Row 1 = week 2 (0-indexed)
print("Week 2 sales   :", weekly[1])           # [170 230 190 140]

# flatten() always returns a copy as a 1-D array
flat = weekly.flatten()
print("Flattened shape:", flat.shape)          # (12,)

# ── WHY ────────────────────────────────────────────────────────────────────
# reshape(-1, 4) is a shortcut: -1 tells NumPy to compute that dimension.
# flatten() vs ravel(): flatten() always copies; ravel() returns a view when
# possible — safer to use flatten() if you plan to modify the 1-D result.
```
