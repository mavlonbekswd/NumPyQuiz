Quiz 03 — Shape, ndim, size  [SOLVED]
Theme  : University Marks
Topics : .shape, .ndim, .size, .dtype

```python
import numpy as np

marks = np.array([[72, 85, 60, 91, 55, 78],
                  [88, 74, 69, 95, 82, 71],
                  [60, 77, 83, 68, 90, 65]])

# .shape → (rows, cols) | .ndim → number of axes | .size → total elements
print("Shape :", marks.shape, " | ndim :", marks.ndim,
      " | size :", marks.size, " | dtype :", marks.dtype)

print("Total entries :", marks.size)           # 3 × 6 = 18

# Row 2, col 1 (0-indexed)
print("Mark [course 3, student 2] :", marks[2, 1])   # 77

# Row 1 gives all 6 marks for course 2
print("Course 2 marks :", marks[1])            # [88 74 69 95 82 71]

# ── WHY ────────────────────────────────────────────────────────────────────
# .size == product of all shape dimensions (3 * 6 = 18).
# .ndim counts the number of brackets you'd need to fully nest the array.
# marks[2, 1] is equivalent to marks[2][1] but more efficient (single lookup).
```
