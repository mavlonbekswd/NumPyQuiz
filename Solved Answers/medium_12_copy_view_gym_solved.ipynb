Quiz 32 — Copy vs View (Advanced)  [SOLVED]
Theme  : Gym Membership

```python
import numpy as np

np.random.seed(8)
visits = np.random.randint(0, 30, (10, 6))
original_snapshot = visits.copy()  # save for comparison

# Slice creates a VIEW — shares memory
view = visits[1:5]
print("view.base is visits  :", view.base is visits)   # True

# Fancy indexing (list of indices) always creates a COPY
copy = visits[[1,2,3,4]]
print("copy.base is None    :", copy.base is None)     # True

# Modifying view changes the original
view += 2
print("Original changed (rows 1-4 col 0):", visits[1:5, 0])

# Modifying copy does NOT change original
copy += 5
changed = np.any(visits[1:5] != original_snapshot[1:5] + 2)
print("Original after copy+5 changed:", changed)   # False

# flags.owndata: True if array owns its data (i.e. is a copy)
print("view.flags.owndata   :", view.flags.owndata)  # False
print("copy.flags.owndata   :", copy.flags.owndata)  # True

# ── WHY ────────────────────────────────────────────────────────────────────
# This is the most important NumPy pitfall: slices are views (fast but risky),
# fancy indexing (integer list) is always a copy (safe but uses more memory).
# Use .flags.owndata or .base to inspect — critical when debugging mutations.
```
