```python
import numpy as np

np.random.seed(55)
values  = np.random.uniform(5, 500, 200)
units   = np.random.randint(1, 10, 200)
returns = np.random.choice([0,1], 200, p=[0.85, 0.15])
ratings = np.random.randint(1, 6, 200)

# ── 1. Revenue ───────────────────────────────────────────────────────────────
revenue = values * units
print(f"Total revenue: £{revenue.sum():,.2f}")

# ── 2. Segmentation ──────────────────────────────────────────────────────────
segments = np.select(
    [values < 50, values <= 200],
    ["Low",        "Mid"],
    default="High"
)
for seg in ["Low", "Mid", "High"]:
    mask = segments == seg
    print(f"  {seg}: {mask.sum()} orders  |  Mean revenue: £{revenue[mask].mean():.2f}")

# ── 3. Returned vs not-returned ratings ──────────────────────────────────────
ret_mask    = returns == 1
no_ret_mask = returns == 0
print(f"\nMean rating — Returned: {ratings[ret_mask].mean():.2f}"
      f"  |  Not returned: {ratings[no_ret_mask].mean():.2f}")

# ── 4. Top-10 by revenue ─────────────────────────────────────────────────────
top10_idx = np.argsort(revenue)[-10:][::-1]
print("\nTop-10 orders by revenue:")
for i in top10_idx:
    print(f"  Order {i+1:3d}: £{revenue[i]:7.2f}")

# ── 5. Loyalty discount ──────────────────────────────────────────────────────
loyal_mask     = ratings == 5
discounted_val = values.copy()
discounted_val[loyal_mask] *= 0.95
total_discount = (values[loyal_mask] - discounted_val[loyal_mask]).sum()
print(f"\nLoyalty discount — Affected: {loyal_mask.sum()}  Total: £{total_discount:.2f}")

# ── 6. Percentiles ───────────────────────────────────────────────────────────
for label, arr in [("Before discount", values), ("After discount", discounted_val)]:
    p10, p50, p90 = np.percentile(arr, [10, 50, 90])
    print(f"{label} — P10: £{p10:.2f}  P50: £{p50:.2f}  P90: £{p90:.2f}")

# ── WHY ───────────────────────────────────────────────────────────────────────
# np.select with a list of conditions is cleaner than nested np.where for 3+
# categories.  First matching condition wins — order matters.
# discounted_val[loyal_mask] *= 0.95 modifies only the relevant subset in-place,
# leaving all other orders unchanged.
```
