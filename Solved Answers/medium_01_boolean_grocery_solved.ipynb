Quiz 21 — Boolean Masking + Filtering + Aggregation  [SOLVED]
Theme  : Grocery Sales
Topics : boolean masks, compound conditions, axis stats, np.where

```python
import numpy as np

np.random.seed(7)
sales = np.random.randint(200, 1200, size=(4, 7))
print("Sales grid:\n", sales)

# 1. Overall mean (all 28 values)
overall_mean = np.mean(sales)
print(f"\nMean sales: {overall_mean:.2f}")

# Boolean mask across full 2-D array
above_mask  = sales > overall_mean
print("Above-mean count:", above_mask.sum())
print("Above-mean values:", sales[above_mask])

# 2. Compound condition — both bounds must be True simultaneously
between_mask = (sales >= 500) & (sales <= 800)
print("\nSales 500-800:", sales[between_mask])

# 3. Per-week mean: axis=1 collapses columns, leaving one value per row
week_means = sales.mean(axis=1)
print("\nWeek means    :", np.round(week_means, 1))

labels = np.where(week_means >= 700, "Strong", "Weak")
print("Week labels   :", labels)

# 4. argmax across columns (axis=1) → index of best day in each week
peak_idx = np.argmax(sales, axis=1)
print("Peak day per week (index):", peak_idx)
print("Peak day values           :", sales[np.arange(4), peak_idx])

# ── WHY ────────────────────────────────────────────────────────────────────
# sales[above_mask] works on the flattened boolean mask — NumPy automatically
# handles multi-dimensional masks.
# sales[np.arange(4), peak_idx] uses fancy row-col indexing:
# row 0 → col peak_idx[0], row 1 → col peak_idx[1], etc.
```
