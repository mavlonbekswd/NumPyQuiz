Quiz 12 — Filtering Arrays
Theme  : Football Player Stats
Topics : boolean filtering, compound conditions (&, |)

SCENARIO
Goals scored by 10 players this season:
goals = [5, 12, 3, 18, 7, 22, 9, 15, 1, 11]
Corresponding ages:
ages  = [24, 28, 22, 31, 26, 29, 23, 27, 20, 25]

TASKS
1. Create both arrays.
2. Filter players who scored more than 10 goals.
3. Filter players aged under 27 AND scored more than 5 goals.
4. Filter players who scored fewer than 5 OR more than 15 goals.
5. Print all three filtered results.

EXPECTED OUTPUT
Over 10 goals       : [12 18 22 15 11]
Young + productive  : [12  7 11]
Few or many goals   : [ 3 18 22  1]

```python
import numpy as np

goals = [5, 12, 3, 18, 7, 22, 9, 15, 1, 11]
ages  = [24, 28, 22, 31, 26, 29, 23, 27, 20, 25]

# ── write your solution below ──────────────────────────────────────────────

goals_arr = np.array(goals)
ages_arr = np.array(ages)

Over = goals_arr[goals_arr > 10]
Young_Productive = ages_arr[ages_arr < 27] & goals_arr[goals_arr < 5]
print(Over)
print(Young_Productive[ages])


```
