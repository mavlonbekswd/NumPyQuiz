# np.cumsum is useful for running totals — e.g. tracking budget spend over time.
# np.ptp gives the spread of the data in a single call.

```python
import numpy as np

salaries = [2800, 3200, 4500, 2200, 5100, 3800, 2600, 4200, 3600, 4900]

# ── write your solution below ──────────────────────────────────────────────

arr_sal = np.array(salaries)

lowest = np.min(arr_sal)
highest = np.max(arr_sal)
total = np.sum(arr_sal)
range = np.ptp(arr_sal)
payroll =np.cumsum(arr_sal)

print("Min salary   :",lowest)
print("Max salary   :",highest)
print("Sum salary   :", total)
print("Range salary   :", range)
print("Payroll   :", payroll)
```
