Quiz 01 — NumPy Arrays + Indexing
Theme  : Call Center
Topics : np.array, basic indexing, max

SCENARIO
A call centre records calls received each hour over a 12-hour shift.
calls = [34, 45, 23, 67, 89, 56, 45, 34, 78, 90, 12, 55]

TASKS
1. Create a NumPy array from the list.
2. Print the total calls received in the shift.
3. Print the first-hour and last-hour call counts.
4. Print calls for hours 4 to 8 inclusive (1-indexed → index 3:8).
5. Print the highest call count recorded.

EXPECTED OUTPUT
Total calls : 628
First hour  : 34  |  Last hour : 55
Hours 4-8   : [67 89 56 45 34]
Busiest hr  : 90

```python
import numpy as np

calls = [34, 45, 23, 67, 89, 56, 45, 34, 78, 90, 12, 55]

# ── write your solution below ──────────────────────────────────────────────

print(np.sum(calls))
print("First hours", calls[0],"| " "Second hour", calls[-1] )
print(calls[3:8])
print(np.max(calls))
```
