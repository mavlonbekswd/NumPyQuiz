Quiz 13 — Stacking + Joining
Theme  : Travel Bookings
Topics : np.hstack, np.vstack, np.concatenate

SCENARIO
Two travel agencies report weekly bookings (7 days each):
agency_a = [45, 38, 52, 60, 70, 85, 90]
agency_b = [30, 42, 55, 48, 65, 78, 88]

TASKS
1. Create both arrays.
2. Concatenate them into one 14-element array (agency A first).
3. Stack them vertically (vstack) into a (2, 7) matrix.
4. Stack them horizontally (hstack) into a (14,) array — same as step 2.
5. Print the total combined bookings for each day using the vstack result.

EXPECTED OUTPUT
Combined (14,) : [45 38 52 60 70 85 90 30 42 55 48 65 78 88]
Vstack shape   : (2, 7)
Daily totals   : [75 80 107 108 135 163 178]

| Funksiya      | Ishlaydi     | Moslashuv        |
| ------------- | ------------ | ---------------- |
| `hstack`      | yonma-yon    | faqat horizontal |
| `vstack`      | ustma-ust    | faqat vertical   |
| `concatenate` | ikkalasi ham | eng flexible     |


```python
import numpy as np

agency_a = [45, 38, 52, 60, 70, 85, 90]
agency_b = [30, 42, 55, 48, 65, 78, 88]

# ── write your solution below ──────────────────────────────────────────────
arr_a = np.array(agency_a)
arr_b = np.array(agency_b)

conct = np.concatenate((arr_a, arr_b))

print(conct)

```
