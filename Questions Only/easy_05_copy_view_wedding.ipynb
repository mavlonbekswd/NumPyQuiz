Quiz 05 — Copy vs View
Theme  : Wedding Guests
Topics : view (slice), copy(), id(), base

SCENARIO
Wedding guest ages (years):
ages = [22, 35, 28, 45, 30, 52, 18, 61, 40, 33]

TASKS
1. Create the NumPy array.
2. Create a VIEW of the first 5 guests (slice).
3. Create a COPY of the last 5 guests.
4. Modify view[0] = 99. Print the original array — does it change?
5. Modify copy[0] = 99. Print the original array — does it change?
6. Use .base to confirm which is a view and which is a copy.

EXPECTED OUTPUT
After view change  : [99 35 28 45 30 52 18 61 40 33]  ← original changed
After copy change  : [99 35 28 45 30 52 18 61 40 33]  ← original unchanged
view.base is ages  : True
copy.base is None  : True

Slice / View
Vazifasi: Arrayning bir qismini oladi. Asl array bilan bog'langan bo'ladi.

copy()
Vazifasi: Arrayning mustaqil nusxasini yaratadi.

base
Vazifasi: Array qayerdan kelganini ko'rsatadi.

```python
import numpy as np

ages = [22, 35, 28, 45, 30, 52, 18, 61, 40, 33]

# ── write your solution below ──────────────────────────────────────────────

arr = np.array(ages)
#print(arr)

view = arr[0:5].view()
copy = arr[5:10].copy()
view[1] = 12
copy[1] = 99
#print(view)
#print(copy)
print(arr)
#print(view.base)
#print(copy.base)
```
