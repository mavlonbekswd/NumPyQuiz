Quiz 08 — 2-D Slicing
Theme  : E-commerce Orders
Topics : 2-D slicing, axis-wise sum, column extraction

SCENARIO
An online store tracks orders across 5 product categories for 4 weeks.
Rows = weeks, Columns = [Electronics, Clothing, Books, Food, Toys]
orders = [[200,150,80,300,120],
           [180,200,90,280,140],
           [220,170,70,320,160],
           [190,160,85,310,130]]

TASKS
1. Create the array.
2. Print total orders per week (sum across columns, axis=1).
3. Print total orders per category (sum across rows, axis=0).
4. Extract only the Electronics and Clothing columns (cols 0 and 1) for all weeks.
5. Print the orders for weeks 2 and 3 only (rows 1 and 2).

EXPECTED OUTPUT
Orders per week    : [ 850  890  940  875]
Orders per category: [790 680 325 1210 550]
Electronics+Cloth  :
[[200 150]  [180 200]  [220 170]  [190 160]]
Weeks 2-3   :
[[180 200  90 280 140]  [220 170  70 320 160]]

```python
import numpy as np

orders = [[200,150,80,300,120],
          [180,200,90,280,140],
          [220,170,70,320,160],
          [190,160,85,310,130]]

# ── write your solution below ──────────────────────────────────────────────

arr_orders = np.array(orders)

per_week = np.sum(arr_orders, axis=1)
per_cat = np.sum(arr_orders, axis=0)

print(per_week)
print(per_cat)

#cloth = arr_orders[[0:,1 0:,0]]
elect = arr_orders[0:1, 0:1]
#extract = cloth elect
#print(cloth)
print(elect)
#print(extract)
```
