Quiz 14 — Random randint + Seed
Theme  : Customer Support Tickets
Topics : np.random.seed, np.random.randint, basic stats on random data

SCENARIO
Simulate a week (7 days) of customer support tickets received per day.
Tickets range from 20 to 80 per day.

TASKS
1. Set random seed to 42 for reproducibility.
2. Generate a random array of 7 integers between 20 and 80 (inclusive).
3. Print the simulated ticket counts.
4. Print the busiest day (max), quietest day (min), and average tickets/day.
5. Print the total tickets for the week.

EXPECTED OUTPUT (seed=42)
Tickets/day : [52 55 47 61 64 25 32]  ← may vary with NumPy version
Busiest     : 64  |  Quietest : 25
Average     : 48.0
Weekly total: 336

```python
import numpy as np

# ── write your solution below ──────────────────────────────────────────────
np.random.seed(42)
arr = (np.random.randint(20, 81,size=7))
print("Busiest:", np.max(arr), "| Quietest:", np.min(arr))
avg = round(np.average(arr),2)

print(arr)

print(avg)

print(np.sum(arr))
```
