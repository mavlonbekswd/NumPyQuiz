SCENARIO
Final exam scores for 10 students (pass mark = 50):
scores = [72, 45, 83, 38, 90, 55, 29, 61, 48, 77]

TASKS
1. Create the array.
2. Use np.where to create a string label array: "Pass" or "Fail".
3. Use np.where to give each student a grade bonus:
   score >= 70 → add 5 bonus points; otherwise add 0.
4. Print labels and adjusted scores.
5. Count how many students passed.

EXPECTED OUTPUT
Labels  : ['Pass' 'Fail' 'Pass' 'Fail' 'Pass' 'Pass' 'Fail' 'Pass' 'Fail' 'Pass']
Adjusted: [77 45 88 38 95 55 29 66 48 82]

```python
import numpy as np

scores = [72, 45, 83, 38, 90, 55, 29, 61, 48, 77]

# ── write your solution below ──────────────────────────────────────────────

scores_arr = np.array(scores)

passmark = scores_arr > 50

bonus = np.where(scores_arr >= 70, "Garde 5",  "0")

print(np.where(passmark, "Pass", "Fail"))

print("Adjust: ", np.where(scores_arr >= 70,  ))

print(bonus)
```
